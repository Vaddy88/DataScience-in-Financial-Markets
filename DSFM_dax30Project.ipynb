```python
import yfinance as yf
import pandas as pd


# Define the date range
start_date = '2014-10-11'
end_date = '2023-10-11'

# Define a dictionary mapping stock symbols to their sectors
dax_stock_sectors = {
    'ZAL.DE': 'CD',
    'BEI.DE': 'ID',
    'RWE.DE': 'UT',
    'HEI.DE': 'ID',
    'P911.DE': 'N/A',
    'BAS.DE': 'MT',
    'DTE.DE': 'UT',
    'DHL.DE': 'ID',
    'FRE.DE': 'FN',
    'SY1.DE': 'ID',
    '1COV.DE': 'MT',
    'BMW.DE': 'CD',
    'MTX.DE': 'MT',
    'DTG.DE': 'ID',
    'HNR1.DE': 'EN',
    'VOW3.DE': 'CD',
    'VNA.DE': 'HC',
    'SHL.DE': 'HC',
    'EOAN.DE': 'UT',
    'ADS.DE': 'IT',
    'DBK.DE': 'FN',
    'BAYN.DE': 'HC',
    'ALV.DE': 'FN',
    'CON.DE': 'ID',
    'DB1.DE': 'FN',
    'SIE.DE': 'ID',
    'IFX.DE': 'IT',
    'AIR.DE': 'ID',
    'MRK.DE': 'HC',
    'ENR.DE': 'EN'
}

# Create an empty DataFrame to store the stock data
dax_data = pd.DataFrame()

# Loop through each stock in the DAX index and fetch historical data
for stock, sector in dax_stock_sectors.items():
    data = yf.download(stock, start=start_date, end=end_date)
    dax_data[sector + '_' + stock] = data['Adj Close']

# Display the stock data
print(dax_data.head())
```

```python
import numpy as np
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

# Assuming you have a DataFrame 'df' with stocks as columns and dates as index
# Make sure your DataFrame contains the stock names as a list or array
# You can extract the stock names like this:
stock_names = grouped_data_cleaned.columns.tolist()
stock_data = grouped_data_cleaned.to_numpy()  # 3652 days over 10 years, first 100 stocks

# Create a graph to build the MST
G = nx.Graph()

# Add nodes to the graph (the first 100 stocks) with labels
num_stocks = stock_data.shape[1]
node_labels = {i: stock_names[i] for i in range(num_stocks)}  # Stock name labels
G.add_nodes_from(range(num_stocks))

# Calculate the Euclidean distance between stocks based on their closing prices
for i in range(num_stocks):
    for j in range(i + 1, num_stocks):
        dist = np.linalg.norm(stock_data[:, i] - stock_data[:, j])
        G.add_edge(i, j, weight=dist)

# Calculate the minimum spanning tree
mst = nx.minimum_spanning_tree(G)

# Use a different layout algorithm (e.g., Kamada-Kawai layout) for better spacing
pos = nx.kamada_kawai_layout(G)

# Adjust node size and edge width
node_size = 30
edge_width = 2

plt.figure(figsize=(18, 15))
nx.draw_networkx_nodes(G, pos, node_size=node_size)
nx.draw_networkx_edges(G, pos, edgelist=mst.edges(), width=edge_width, alpha=0.5)
nx.draw_networkx_labels(G, pos, labels=node_labels, font_size=8, font_color='black', verticalalignment='bottom')
plt.axis('off')
plt.title('MST of the First 30 Stocks with Improved Layout')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Calculate daily returns
rolling_window = 20  # You can adjust the window size as needed
returns_data = grouped_data_cleaned.copy()
rv_data = grouped_data_cleaned.copy()
for col in returns_data.columns:
  returns_data[col] = returns_data[col].pct_change().bfill()
  rv_data[col] = returns_data[col].rolling(window=rolling_window).std().fillna(0)
```

```python
import pandas as pd
import numpy as np

grouped_data_cleaned.index = pd.to_datetime(grouped_data_cleaned.index)

# Calculate SDR (Standard Deviation Ratio) for each year and month
def calculate_sdr(data, time_horizon):
    if time_horizon == 'yearly':
        resampled_data = data.groupby(grouped_data_cleaned.index.year).last()  # Resample to yearly data
    elif time_horizon == 'monthly':
        resampled_data = data.groupby([grouped_data_cleaned.index.year, grouped_data_cleaned.index.month]).last()  # Resample to monthly data
        resampled_data.index = resampled_data.index.get_level_values(1).astype(str) + '/' + resampled_data.index.get_level_values(0).astype(str)
    else:
        raise ValueError("Invalid time_horizon, use 'yearly' or 'monthly'.")

    returns = resampled_data.pct_change().dropna()  # Calculate returns
    sdr = returns.std(axis = 1) / returns.mean(axis = 1)
    return sdr

# Calculate VIX (Volatility Index) for each year and month
def calculate_vix(data, time_horizon):
    if time_horizon == 'yearly':
        resampled_data = data.groupby(grouped_data_cleaned.index.year).last()  # Resample to yearly data
    elif time_horizon == 'monthly':
        resampled_data = data.groupby([grouped_data_cleaned.index.year, grouped_data_cleaned.index.month]).last()  # Resample to monthly data
        resampled_data.index = resampled_data.index.get_level_values(1).astype(str) + '/' + resampled_data.index.get_level_values(0).astype(str)
    else:
        raise ValueError("Invalid time_horizon, use 'yearly' or 'monthly'.")

    returns = resampled_data.pct_change().dropna()  # Calculate returns
    vix = returns.std(axis = 1)
    return vix

# Load your stock price data into a DataFrame

# Calculate yearly and monthly SDR
yearly_sdr = calculate_sdr(grouped_data_cleaned, 'yearly')
monthly_sdr = calculate_sdr(grouped_data_cleaned, 'monthly')

# Calculate yearly and monthly VIX
yearly_vix = calculate_vix(grouped_data_cleaned, 'yearly')
monthly_vix = calculate_vix(grouped_data_cleaned, 'monthly')

print("Yearly SDR:")
print(yearly_sdr)
print("\nMonthly SDR:")
print(monthly_sdr)

print("\nYearly VIX:")
print(yearly_vix)
print("\nMonthly VIX:")
print(monthly_vix)
```
