```python
import numpy as np
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

# Assuming you have a DataFrame 'ftse_cleaned_data' with stocks as columns and dates as index
# Make sure your DataFrame contains the stock names as a list or array
# You can extract the stock names like this:
stock_names = ftse_cleaned_data.columns.tolist()
stock_data = ftse_cleaned_data.to_numpy()  # 3652 days over 10 years, first 100 stocks

# Check for missing values in the DataFrame
missing_values = np.isnan(stock_data)

# Handle missing values, e.g., by filling them with zeros
stock_data = np.nan_to_num(stock_data, nan=0)

# Create a graph to build the MST
G = nx.Graph()

# Add nodes to the graph (the first 100 stocks) with labels
num_stocks = stock_data.shape[1]
node_labels = {i: stock_names[i] for i in range(num_stocks)}  # Stock name labels
G.add_nodes_from(range(num_stocks))

# Calculate the Euclidean distance between stocks based on their closing prices
for i in range(num_stocks):
    for j in range(i + 1, num_stocks):
        dist = np.linalg.norm(stock_data[:, i] - stock_data[:, j])
        G.add_edge(i, j, weight=dist)

# Calculate the minimum spanning tree
mst = nx.minimum_spanning_tree(G)

# Use a different layout algorithm (e.g., Kamada-Kawai layout) for better spacing
pos = nx.kamada_kawai_layout(G)

# Adjust node size and edge width
node_size = 50
edge_width = 2

plt.figure(figsize=(10, 10))
nx.draw_networkx_nodes(G, pos, node_size=node_size)
nx.draw_networkx_edges(G, pos, edgelist=mst.edges(), width=edge_width, alpha=0.5)
nx.draw_networkx_labels(G, pos, labels=node_labels, font_size=8, font_color='black', verticalalignment='bottom')
plt.axis('off')
plt.title('MST of the First 100 Stocks with Improved Layout')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Calculate daily returns
rolling_window = 20  # You can adjust the window size as needed
returns_data = ftse_cleaned_data.copy()
rv_data = ftse_cleaned_data.copy()
for col in returns_data.columns:
  returns_data[col] = returns_data[col].pct_change().bfill()
  rv_data[col] = returns_data[col].rolling(window=rolling_window).std().fillna(0)
```

```python
import pandas as pd
import numpy as np

ftse_cleaned_data.index = pd.to_datetime(ftse_cleaned_data.index)

# Calculate SDR (Standard Deviation Ratio) for each year and month
def calculate_sdr(data, time_horizon):
    if time_horizon == 'yearly':
        resampled_data = data.groupby(ftse_cleaned_data.index.year).last()  # Resample to yearly data
    elif time_horizon == 'monthly':
        resampled_data = data.groupby([ftse_cleaned_data.index.year, ftse_cleaned_data.index.month]).last()  # Resample to monthly data
        resampled_data.index = resampled_data.index.get_level_values(1).astype(str) + '/' + resampled_data.index.get_level_values(0).astype(str)
    else:
        raise ValueError("Invalid time_horizon, use 'yearly' or 'monthly'.")

    returns = resampled_data.pct_change().dropna()  # Calculate returns
    sdr = returns.std(axis = 1) / returns.mean(axis = 1)
    return sdr

# Calculate VIX (Volatility Index) for each year and month
def calculate_vix(data, time_horizon):
    if time_horizon == 'yearly':
        resampled_data = data.groupby(ftse_cleaned_data.index.year).last()  # Resample to yearly data
    elif time_horizon == 'monthly':
        resampled_data = data.groupby([ftse_cleaned_data.index.year, ftse_cleaned_data.index.month]).last()  # Resample to monthly data
        resampled_data.index = resampled_data.index.get_level_values(1).astype(str) + '/' + resampled_data.index.get_level_values(0).astype(str)
    else:
        raise ValueError("Invalid time_horizon, use 'yearly' or 'monthly'.")

    returns = resampled_data.pct_change().dropna()  # Calculate returns
    vix = returns.std(axis = 1)
    return vix

# Load your stock price data into a DataFrame

# Calculate yearly and monthly SDR
yearly_sdr = calculate_sdr(ftse_cleaned_data, 'yearly')
monthly_sdr = calculate_sdr(ftse_cleaned_data, 'monthly')

# Calculate yearly and monthly VIX
yearly_vix = calculate_vix(ftse_cleaned_data, 'yearly')
monthly_vix = calculate_vix(ftse_cleaned_data, 'monthly')

print("Yearly SDR:")
print(yearly_sdr)
print("\nMonthly SDR:")
print(monthly_sdr)

print("\nYearly VIX:")
print(yearly_vix)
print("\nMonthly VIX:")
print(monthly_vix)
```
